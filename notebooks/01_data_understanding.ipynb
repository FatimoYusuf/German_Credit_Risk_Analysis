```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

```python
#Loading the dataset
df = pd.read_csv("../data/German_Credit.csv")
```

```python
#Understanding the structure and basic properties of the dataset.
df.head()
df.shape
df.info()
```

The Data Structure:

- Total rows: 1000
- Total columns: 10
- Target variable: `Risk` (1 = good, 0 = bad)
- Feature types:
  - Numeric: Age, Credit amount, Duration, Risk
  - Categorical: Sex, Job, Housing, Saving accounts, Checking account, Purpose

Observation:
- No missing values detected
- Data types are mostly correct
- Some columns have spaces in names (e.g., `Saving accounts`), so it needs to be cleaned by using _ instead of space.

```python
# Missing values
df.isna().sum()
# Duplicates
df.duplicated().sum()

# Statistical summary
df.describe(include='all')
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
df.rename(columns={
    'saving_accounts': 'saving_account'
}, inplace=True)
```

```python
#Check missing values
df.isna().sum().sort_values(ascending=False)
```

```python
#confirming datatypes
df.dtypes
```

```python
df['age'].min(), df['age'].max()
df['duration'].min(), df['duration'].max()
df['credit_amount'].min(), df['credit_amount'].max()
df['risk'].value_counts(normalize=True)
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18, 34, 64, 75],
    labels=['Young', 'Middle-aged', 'Retired'],
    include_lowest=True
)
```

```python
df['credit_per_month'] = df['credit_amount'] / df['duration']
```

```python
# df['has_savings'] = df['saving_accounts'].apply(lambda x: 'Yes' if x != 'none' else 'No')
# df['has_checking'] = df['checking_account'].apply(lambda x: 'Yes' if x != 'none' else 'No')


saving_map = {'none': 0, 'little': 1, 'moderate': 2, 'rich': 3, 'quite rich': 4}
checking_map = {'none': 0, 'little': 1, 'moderate': 2, 'rich': 3, 'quite rich': 4}

df['saving_score'] = df['saving_account'].map(saving_map)
df['checking_score'] = df['checking_account'].map(checking_map)
```
